```python
Weight = int(input("Enter the BMI Index:"))
if (Weight<16):
    print("Severe Thinness")
elif (Weight<18.5):
    print("Mild Thinness")
elif (Weight<25):
    print("Normal")
elif (Weight<30):
    print("Overweight")
elif (Weight<40):
    print("Very Overweight")
else:
    print("Obese Class II")
```
